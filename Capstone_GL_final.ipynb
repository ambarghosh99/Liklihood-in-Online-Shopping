```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Revenue'] = le.fit_transform(data['Revenue'])
data['Revenue'].value_counts()
```

```python
data1['Weekend'] = le.fit_transform(data1['Weekend'])
data1['Weekend'].value_counts()
```

```python
x = data1
# removing the target column revenue from x
x = x.drop(['Revenue'], axis = 1)

y = data['Revenue']

# checking the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
data1.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1, stratify=y)

# checking the shapes

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix, classification_report, precision_recall_curve
from sklearn.linear_model import LogisticRegression

m1 = LogisticRegression()
m1.fit(x_train,y_train)
ypred = m1.predict(x_test)
print("Accuracy metric is:",accuracy_score(y_test,ypred))
print("Confusion metrics is:\n",confusion_matrix(y_test,ypred))

cm=confusion_matrix(y_test,ypred)
cr=classification_report(y_test,ypred)

specificity=cm[0,0]/cm[0,:].sum()
print("Specificity or the power or true Negative rate or Recall Healthy or 1-alpha or tnr:",specificity)

sensitivity=cm[1,1]/cm[1,:].sum()
print("Sensitivity or the power or true positive rate or Recall Disease or 1-Beta or tpr is:",sensitivity)

precision_0 = cm[0,0]/cm[:,0].sum()
precision_0

precision_1 = cm[1,1]/cm[:,1].sum()
precision_1
print(cr)
```

```python
ypred_new=[]
for y in m1.predict_proba(x_test)[:,1]:
  #  if round(y,2)>=0.43: # Checking the value, tried with mean, median and random
    if y>=0.3:
       # print(y)
        ypred_new.append(1)
    else:
        #print(y)
        ypred_new.append(0)
```

```python
print(confusion_matrix(y_test,ypred_new))
print(classification_report(y_test,ypred_new))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,KFold,cross_val_score,LeaveOneOut 
```

```python
models = []
models.append(('LR', LogisticRegression()))
#models.append(('LDA', LinearDiscriminantAnalysis()))

models.append(('CART', DecisionTreeClassifier()))

models.append(('RFC', RandomForestClassifier()))
#models.append(('SVM', SVC(gamma='auto')))
# evaluate each model in turn
results = []
names = []
scoring = 'recall' #accuracy, f1_weighted, 
for name, model in models:
	kfold = KFold(n_splits=10, random_state=12345)
	cv_results = cross_val_score(model,x_train, y_train, cv=kfold, scoring='recall')
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)

# Train the classifier
clf.fit(x_train, y_train)

# Print the name and gini importance of each feature
for feature in zip(x_train.columns, clf.feature_importances_):
    print(feature)
```

```python
# Create a selector object that will use the random forest classifier to identify
# features that have an importance of more than 0.01
sfm = SelectFromModel(clf, threshold=0.01)

# Train the selector
sfm.fit(x_train, y_train)
```

```python
# Print the names of the most important features
important_cols=[]
for feature_list_index in sfm.get_support(indices=True):
    important_cols.append(x_train.columns[feature_list_index])
    print(x_train.columns[feature_list_index])
```

```python
# Transform the data to create a new dataset containing only the most important features
# Note: We have to apply the transform to both the training X and test X data.
X_important_train = sfm.transform(x_train)
X_important_test = sfm.transform(x_test)
```

```python
# Create a new random forest classifier for the most important features
clf_important = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)

# Train the new classifier on the new dataset containing the most important features
clf_important.fit(X_important_train, y_train)
```

```python
# Apply The Full Featured Classifier To The Test Data
y_pred = clf.predict(x_test)

# View The Accuracy Of Our Full Feature (4 Features) Model
accuracy_score(y_test, y_pred)
```

```python
# Apply The Full Featured Classifier To The Test Data
y_important_pred = clf_important.predict(X_important_test)

# View The Accuracy Of Our Limited Feature (2 Features) Model
accuracy_score(y_test, y_important_pred)
```

```python
model = RandomForestClassifier(n_estimators=1000)
model.fit(X_important_train, y_train)

y_pred = model.predict(X_important_test)

# evaluating the model
print("Training Accuracy :", model.score(X_important_train, y_train))
print("Testing Accuracy :", accuracy_score(y_test, y_important_pred))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)
print(cm)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
kfold = model_selection.KFold(n_splits=10)
results = model_selection.cross_val_score(model, x.loc[:,important_cols], y, cv=kfold)
print(results.mean())
```
